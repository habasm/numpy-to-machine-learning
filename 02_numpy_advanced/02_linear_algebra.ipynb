### 3. Dot Product

```python
a = np.array([1, 2])
b = np.array([3, 4])

dot = np.dot(a, b)
print("Dot product:", dot)
```

### 4. Matrix Multiplication

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

result = np.matmul(A, B)
print(result)
```

### 5. Transpose

```python
print(A.T)
```

### 6. Determinant

```python
det = np.linalg.det(A)
print("Determinant:", det)
```

### 7. Inverse

```python
inv = np.linalg.inv(A)
print(inv)
```
