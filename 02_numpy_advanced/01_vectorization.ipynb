### 1. Vectorization in NumPy
Goal:
- Replace loops with fast NumPy operations
- Understand performance benefits


### 2. Imports

```python
import numpy as np
import time
```

### 3. Slow (Loop Version)

```python
arr = np.arange(1_000_000)

start = time.time()

result = []
for x in arr:
    result.append(x * 2)

end = time.time()

print("Loop time:", end - start)
```

### 4. Fast (Vectorized)

```python
start = time.time()

result = arr * 2

end = time.time()

print("Vectorized time:", end - start)
```

### 5. Comparison Insight

Vectorization is:
- Faster
- Cleaner
- Essential for ML computations

### 6. Real ML Example (Loss Calculation)

```python
y_true = np.array([1, 0, 1, 1])
y_pred = np.array([0.9, 0.2, 0.8, 0.7])

# Mean Squared Error
mse = np.mean((y_true - y_pred) ** 2)

print("MSE:", mse)
```
