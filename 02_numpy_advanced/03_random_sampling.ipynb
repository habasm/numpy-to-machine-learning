### 1.  Random Sampling in NumPy

### 2. Imports

```python
import numpy as np
```

### 3. Random Numbers

```python
print(np.random.rand(3))       # uniform
print(np.random.randn(3))      # normal
```

### 4. Random Integers

```python
print(np.random.randint(0, 10, size=5))
```

### 5. Shuffle

```python
arr = np.array([1, 2, 3, 4, 5])
np.random.shuffle(arr)

print(arr)
```

### 6. Sampling

```python
sample = np.random.choice([10, 20, 30, 40], size=3)
print(sample)
```

### 7. Reproducibility

```python
np.random.seed(42)

print(np.random.rand(3))
```
