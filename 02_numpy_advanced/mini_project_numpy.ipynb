### 1.  Mini Project: Data Analysis with NumPy

### 2. Setup

```python
import numpy as np
```

### 3. Create Dataset

```python
# Simulated student scores
scores = np.random.randint(40, 100, size=50)

print(scores)
```

### 4. Basic Analysis

```python
print("Mean:", scores.mean())
print("Max:", scores.max())
print("Min:", scores.min())
```

### 5. Standard Deviation

```python
print("Std Dev:", scores.std())
```

### 6. Top Performers

```python
top_students = scores[scores > 85]
print(top_students)
```

### 7. Grade Classification

```python
grades = []

for s in scores:
    if s >= 75:
        grades.append("Pass")
    else:
        grades.append("Fail")

print(grades[:10])
```

### 8. Vectorized Version (Better)

```python
grades = np.where(scores >= 75, "Pass", "Fail")
print(grades[:10])
```
