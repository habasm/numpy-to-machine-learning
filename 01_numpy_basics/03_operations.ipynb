### 1. Arithmetic

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)
print(a * b)
```

### 2. Scalar Ops

```python
print(a + 10)
print(a * 2)
```

### 3. Aggregations

```python
print("Sum:", a.sum())
print("Mean:", a.mean())
print("Max:", a.max())
```
