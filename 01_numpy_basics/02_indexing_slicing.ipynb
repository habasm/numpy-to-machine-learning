### Core Idea

#### Indexing & Slicing

### 1. Basic Indexing

```python
import numpy as np

arr = np.array([10, 20, 30, 40])

print(arr[0])
print(arr[-1])
```

### 2. Slicing

```python
print(arr[1:3])
print(arr[:2])
```

### 3. 2D Indexing

```python
mat = np.array([[1, 2], [3, 4]])

print(mat[0, 1])  # 2
```

### 4. Boolean Indexing

```python
arr = np.array([1, 2, 3, 4, 5])

filtered = arr[arr > 2]
print(filtered)
```

5. Practice

```python
# Extract all values > 3 from:
arr = np.array([1, 3, 5, 7, 2])

result = arr[arr > 3]
print(result)
```
