### 1. NumPy Arrays
Learn:
- What arrays are
- How to create them

### 2. Create Arrays

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([[1, 2], [3, 4]])

print(a)
print(b)
```

### 3. Array Properties

```python
print("Shape:", b.shape)
print("Dimensions:", b.ndim)
print("Data type:", b.dtype)
```

### 4. Special Arrays

```python
zeros = np.zeros((2, 3))
ones = np.ones((2, 2))
identity = np.eye(3)

print(zeros)
print(ones)
print(identity)
```
