### 3. Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

### 4. Dataset

```python
X = np.array([[1], [2], [3], [4]])
y = np.array([2, 4, 6, 8])
```

### 5. Train Model

```python
model = LinearRegression()
model.fit(X, y)
```

### 6. Predictions

```python
pred = model.predict([[5]])
print(pred)
```

### 7. Visualization

```python
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.title("Linear Regression Fit")
plt.show()
```

### 8. Insight

Model learned:
y ≈ 2x

It found the best line fitting the data
