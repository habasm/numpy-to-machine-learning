### 1. Vectors for Machine Learning
Goal:
- Understand vectors as data
- Perform basic vector operations

### 2. Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 3. What is a Vector?

```python
v = np.array([2, 3])
print(v)
```

👉 Think of a vector as:

- a point in space
- or a list of features (ML perspective)

### 4. Visualizing a Vector

```python
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1)
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.grid()
plt.title("Vector Visualization")
plt.show()
```

### 5. Vector Addition

```python
v1 = np.array([1, 2])
v2 = np.array([3, 1])

print(v1 + v2)
```

👉 ML meaning:

- combining features

### 6. Scalar Multiplication

```python
v = np.array([2, 3])
print(v * 2)
```

### 7. Magnitude (Length)




```python
v = np.array([3, 4])
mag = np.linalg.norm(v)
print(mag)
```

### 8. Dot 



```python
a = np.array([1, 2])
b = np.array([3, 4])

print(np.dot(a, b))
```

👉 ML meaning:

- similarity between vectors
- core of linear regression
