# Statistics for Machine Learning

### 2. Dataset

```python
import numpy as np
data = np.array([10, 20, 30, 40, 50])
```

### 3. Mean


```python
print(np.mean(data))
```

### 4. Variance


```python
print(np.var(data))
```

### 5. Standard Deviation

```python
print(np.std(data))
```

### 6. Normal Distribution

```python
import matplotlib.pyplot as plt

x = np.random.randn(1000)

plt.hist(x, bins=30)
plt.title("Normal Distribution")
plt.show()
```

### 7. Correlation

```python
x = np.array([1,2,3,4])
y = np.array([2,4,6,8])

print(np.corrcoef(x, y))
```
