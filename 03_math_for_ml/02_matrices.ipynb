### 1. Matrices for Machine Learning

### 2. What is a Matrix?

```python
import numpy as np

X = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])

print(X)
```

👉 Think:

- rows = samples
- columns = features

### 3. Transpose

```python
print(X.T)
```

### 4. Matrix Multiplication

```python
A = np.array([[1, 2]])
B = np.array([[3], [4]])

print(np.matmul(A, B))
```

👉 ML meaning:

- predictions = X · weights

### 5. System of Equations

Ax = b

```python
A = np.array([[2, 1], [1, 3]])
b = np.array([5, 6])

x = np.linalg.solve(A, b)
print(x)
```

### 6. Determinant

👉 tells if matrix is invertible

```python
A = np.array([[1, 2], [3, 4]])
print(np.linalg.det(A))
```

### 7. Inverse

```python
print(np.linalg.inv(A))
```
