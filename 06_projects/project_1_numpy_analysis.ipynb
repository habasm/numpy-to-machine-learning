### Project 1: NumPy Data Analysis

Goal:
- Analyze dataset using NumPy only
- Apply vectorization + statistics

### 2. Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 3. Dataset (Simulated Sales Data)

```python
np.random.seed(42)

sales = np.random.randint(100, 1000, size=100)
print(sales[:10])
```

### 4. Basic Analysis

```python
print("Mean:", sales.mean())
print("Max:", sales.max())
print("Min:", sales.min())
print("Std:", sales.std())
```

### 5. Visualization

```python
plt.hist(sales, bins=20)
plt.title("Sales Distribution")
plt.show()
```

### 6. Insights

```python
high_sales = sales[sales > 800]
print("High sales count:", len(high_sales))
```

### 7. Vectorized Logic

```python
performance = np.where(sales > 500, "Good", "Low")
print(performance[:10])
```

### 8. Final Insight

- Identified high-performing sales
- Used vectorization for classification
- Applied statistical analysis
