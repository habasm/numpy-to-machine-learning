### 1. Goal:
- Handle missing values
- Fix inconsistent data
- Prepare dataset for ML

### 2. Imports

```python
import numpy as np
import pandas as pd
```

### 3. Create Sample Dataset

```python
data = {
    "age": [25, 30, None, 22, 40],
    "salary": [50000, 60000, 55000, None, 65000],
    "city": ["Addis", "Adama", "Addis", None, "Bahir Dar"]
}

df = pd.DataFrame(data)
print(df)
```

### 4. Detect Missing Values

```python
print(df.isnull())
print(df.isnull().sum())
```

### 5. Handle Missing Values

```python
### Option 1: Drop


df_drop = df.dropna()
print(df_drop)

### Option 2: Fill (Better)

df["age"].fillna(df["age"].mean(), inplace=True)
df["salary"].fillna(df["salary"].median(), inplace=True)
df["city"].fillna("Unknown", inplace=True)

print(df)
```

### 6. Remove Duplicates

```python
df = df.drop_duplicates()
```

### 7. Fix Data Types

```python
df["age"] = df["age"].astype(int)
print(df.dtypes)
```
