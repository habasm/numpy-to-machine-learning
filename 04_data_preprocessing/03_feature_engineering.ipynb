###  Feature Engineering

### 2. Imports

```python
import pandas as pd
```

### 3. Sample Data

```python
data = {
    "height": [150, 160, 170],
    "weight": [50, 60, 70]
}

df = pd.DataFrame(data)
```

### 4. Create New Feature (BMI)

```python
df["bmi"] = df["weight"] / (df["height"] / 100) ** 2
print(df)
```

### 5. Encoding Categorical Data

```python
df["city"] = ["Addis", "Adama", "Addis"]

encoded = pd.get_dummies(df["city"])
print(encoded)
```

### 6. Feature Selection

```python
# Drop irrelevant columns
df = df.drop(columns=["city"])
```

### 7. Binning

```python
df["age"] = [18, 25, 40]

df["age_group"] = pd.cut(df["age"],
                        bins=[0, 20, 30, 50],
                        labels=["Young", "Adult", "Senior"])

print(df)
```
