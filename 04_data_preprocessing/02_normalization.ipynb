### # Data Normalization

### 2. Imports

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

### 3. Sample Data

```python
data = np.array([[100], [200], [300], [400]])
```

### 4. Min-Max Scaling



```python
scaler = MinMaxScaler()
scaled = scaler.fit_transform(data)

print(scaled)
```

### 5. Standardization (Z-score)



```python
scaler = StandardScaler()
standardized = scaler.fit_transform(data)

print(standardized)
```

### 6. When to Use

MinMax:
- bounded data
- neural networks

StandardScaler:
- most ML models
- normal distribution assumption
